```python
def extract_column_to_df(df, column_name):
    # Inicializar una lista para almacenar los diccionarios
    extracted_data = []

    # Iterar sobre las filas del DataFrame
    for _, row in df.iterrows():
        # Se extrae información de la columna especificada
        column_info = row[column_name]

        # Verificar si la columna contiene información válida antes de intentar agregar a la lista
        if isinstance(column_info, list) or isinstance(column_info, dict):
            extracted_data.extend(column_info if isinstance(column_info, list) else [column_info])

    # Crear un nuevo DataFrame con el nombre especificado
    new_df_name = f"{column_name}_df"
    globals()[new_df_name] = pd.DataFrame(extracted_data).drop_duplicates().set_index('id')


extract_column_to_df(games_df, 'genres')
```

```python
# FUNCION PARA EXTRAER LAS 

def extract_and_merge(df, column,fila):

    # Lista para almacenar los diccionarios
    platforms_data = []

    # Iterar sobre las filas del DataFrame
    for _, row in df.iterrows():
        # Se extrae información de la columna especificada
        platforms_info = row[column]

        # Iterar sobre los diccionarios dentro de la lista (extraer y agregar a la lista)
        for platform_dict in platforms_info:
            platform_info = platform_dict[fila]
            platforms_data.append(platform_info)

    # Crear un nuevo DataFrame llamado "platforms_df"
    result_df = pd.DataFrame(platforms_data).drop_duplicates().set_index('id')

    return result_df



result_df = extract_and_merge(games_df, 'platforms','platform')
platforms_df= result_df

platforms_df
```

```python
def extract_ids(platforms_list):
    return [platform['platform']['id'] for platform in platforms_list]

# Aplica la función a la columna 'platforms' y crear una nueva columna 'platform_ids'
games_df['platform_ids'] = games_df['platforms'].apply(extract_ids)

# Separa cada registro por las plataformas que esta disponible
games_df = games_df.explode('platform_ids')
games_df

# Imprimir el DataFrame resultante
games_df.head()
```
